```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,n_classes=2, n_clusters_per_class=1, random_state = 42, hypercube=False)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:, 1], c = y, cmap = 'winter', s = 100)
```

```python
def step(z):
  return 1 if z > 0 else 0
```

```python
def perceptron(X,y):

  X = np.insert(X, 0,1,axis=1)  # bias
  weights = np.ones(X.shape[1])
  lr = 0.1

  for i in range(4000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j], weights))
    weights = weights + lr * (y[j] - y_hat) * X[j]

  return (weights[0], weights[1:] )
```

```python
intercept_,coef = perceptron(X, y)
print(coef)
print(intercept_)
```

```python
m = -(coef[0] / coef[1])
b = -(intercept_/coef[1])
```

```python
x_input =  np.linspace(-3, 3, 100)
y_input = m *x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:, 0], X[:,1], c = y, cmap =  'winter', s = 100)
plt.ylim(-3,2)
```

```python
def perceptron_2(X,y):
  m = []
  b = []

  X = np.insert(X, 0,1, axis = 1)
  weights = np.ones(X.shape[1])
  lr = 0.1

  for i in range(200):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j], weights))
    weights = weights + lr*(y[j] - y_hat) * X[j]

    m.append(- (weights[1]/ weights[2]))
    b.append(- (weights[0]/ weights[2]))

  return m , b
```

```python
m ,  b = perceptron_2(X, y)
```

# Perceptron Loss Function

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

X, y = make_classification(n_samples = 200, n_features = 2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube = False, class_sep = 15 )
```

```python
X.shape
```

```python
y
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:, 0], X[:, 1], c = y , cmap = 'winter', s = 100)
```

```python
def perceptro(X, y ):
  w1 = w2 = b = 1
  lr = 0.1

  for j in range(2000):
    for i in range(X.shape[0]):

      z = w1 * X[i][0] + w2 * X[i][1] + b

      if z * y[i] < 0:
        w1  = w1 + lr * y[i] * X[i][0]
        w2  = w2 + lr * y[i] * X[i][1]
        b = b + lr * y[i]

    return w1, w2, b
```

```python
w1, w2, b = perceptro(X, y)
print( "w1: ", w1)
print( "w2: ", w2)
print( "b: ", b)
```

```python
m = -( w1/w2)
c = -(b/w2)
print( "m: ", m)
print( "c: ", c)
```

```python
x_input = np.linspace(-3, 3, 200)
y_input = m * x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color = 'red', linewidth = 3)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap='winter', s = 100)
plt.ylim(-3,3)
```

# A Perceptron class that implements a single-TLU network
`TLU: Threshold Logic Unit (TLU)` == Perceptron( It processes inputs and produces an output based on whether the total input reaches a certain threshold.)

```python
import numpy as np

from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
X = iris.data[:, (2,3)]
y = (iris.target == 0).astype(int)
```

```python
X
```

```python
y
```

```python
perceptron_clf = Perceptron(random_state=42)
perceptron_clf.fit(X,y)
```

```python
y_prediction = perceptron_clf.predict([[2,0.5]])
y_prediction
```
